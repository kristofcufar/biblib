```python
import sys
import biblib.bib
import pandas as pd
import math
```

```python
path = 'DATA\\library.xlsx'
```

```python
def author_reformat(author_str):
    author_list = author_str.split(' and ')
    output = []
    for i in author_list:
        names = i.split('. ')[:-1]
        for j in range(len(names)):
            names[j] += '.'

        surnames = i.split('. ')[-1]
        out = []
        out.append(surnames + ', ' + ' '.join(names))

        output.append(out[0])

    return ' and '.join(output)
```

```python
data = pd.read_excel(path, sheet_name='articles')

articles = ''

for i in range(len(data)):
    article = data.iloc[i]
    articles += '@article{' + article['key'] + ',\n'
    if str(article['author']) != 'nan':
        author_ref = author_reformat(article['author'])
        articles += '  author = {' + author_ref + '},\n'
    if str(article['title']) != 'nan':
        articles += '  title = {' + article['title'] + '},\n'
    if str(article['journal']) != 'nan':
        articles += '  journal = {' + article['journal'] + '},\n'
    if str(article['volume']) != 'nan':
        articles += '  volume = {' + str(article['volume']) + '},\n'
    if str(article['pages']) != 'nan':
        articles += '  pages = {' + str(article['pages']) + '},\n'
    if str(article['year']) != 'nan':
        articles += '  year = {' + str(article['year']) + '},\n'
    if str(article['issue']) != 'nan':
        articles += '  issue = {' + str(article['issue']) + '},\n'
    if str(article['doi']) != 'nan':
        articles += '  doi = {' + '{' + str(article['doi']).replace('_', '\_') + '}' + '},\n'
    articles += '}\n\n'

data = pd.read_excel(path, sheet_name='inproceedings')

inproceedings = ''

for i in range(len(data)):
    inproceeding = data.iloc[i]
    inproceedings += '@inproceedings{' + inproceeding['key'] + ',\n'
    if str(inproceeding['author']) != 'nan':
        inproceedings += '  author = {' + inproceeding['author'] + '},\n'
    if str(inproceeding['title']) != 'nan':
        inproceedings += '  title = {' + inproceeding['title'] + '},\n'
    if str(inproceeding['booktitle']) != 'nan':
        inproceedings += '  booktitle = {' + inproceeding['booktitle'] + '},\n'
    if str(inproceeding['year']) != 'nan':
        inproceedings += '  year = {' + str(inproceeding['year']) + '},\n'
    if str(inproceeding['month']) != 'nan':
        inproceedings += '  month = {' + str(inproceeding['month']) + '},\n'
    if str(inproceeding['pages']) != 'nan':
        inproceedings += '  pages = {' + str(inproceeding['pages']) + '},\n'
    if str(inproceeding['doi']) != 'nan':
        inproceedings += '  doi = {' + '{' + str(inproceeding['doi']).replace('_', '\_') + '}' + '},\n'
    if str(inproceeding['city']) != 'nan':
        inproceedings += '  city = {' + str(inproceeding['city']) + '},\n'
    inproceedings += '}\n\n'

data = pd.read_excel(path, sheet_name='books')

books = ''

for i in range(len(data)):
    book = data.iloc[i]
    books += '@book{' + book['key'] + ',\n'
    if str(book['author']) != 'nan':
        books += '  author = {' + book['author'] + '},\n'
    if str(book['title']) != 'nan':
        books += '  title = {' + book['title'] + '},\n'
    if str(book['publisher']) != 'nan':
        books += '  publisher = {' + book['publisher'] + '},\n'
    if str(book['year']) != 'nan':
        books += '  year = {' + str(book['year']) + '},\n'
    if str(book['city']) != 'nan':
        books += '  city = {' + str(book['city']) + '},\n'
    if str(book['edition']) != 'nan':
        books += '  edition = {' + str(book['edition']) + '},\n'
    books += '}\n\n'


all_entries = articles + inproceedings + books
    
with open('library_2.bib', 'w', encoding='utf8') as f:
    f.write(all_entries)
```
